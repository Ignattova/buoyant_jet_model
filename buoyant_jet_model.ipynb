***Abstract:*** This project investigates the mathematical modeling of key hydrodynamic processes that occur during submarine discharge into coastal waters, where the effluent density differs from that of the ambient water. The primary focus is on the physical processes governing the initial mixing and dilution of the discharged jet, which play a critical role in limiting environmental impacts on marine ecosystems. A one-dimensional (1D) integral plume model is implemented, based on a system of three ordinary differential equations representing the conservation of mass flux (Q), momentum flux (M), and buoyancy flux (B). The model accounts for turbulent entrainment of the ambient water and the effects of vertical density stratification in the seawater, using the Boussinesq approximation. The study provides a theoretical basis for future analysis of the overall mixing processes between wastewater and seawater.

## I. INTRODUCTION

>### I.1 Problem Statement and Motivation
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Coastal regions hold significant environmental and economic value at both global and national scales. The demand for recreation and tourism has led to a high concentration of population in these areas, resulting in rapid urban expansion and increasing pressure on coastal zones. One of the primary sources of this environmental pressure is the discharge of municipal and industrial wastewater.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;*The Bulgarian Black Sea coast* is the main recipient of a significant part of the wastewater in the country - wastewater from urban treatment plants (WWTPs), sewage without complete treatment, as well as significant volumes of urban stormwater runoff. This leads to increased anthropogenic pollution and biogenic nutrient enrichment (eutrophication), which has become a major ecological concern over recent decades. As a result, the entire Bulgarian Black Sea coast has been designated as a ***sensitive area***.


>### I.2 Environmental Quality Standards and Mixing Zones
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The discharge of wastewater of different composition into water bodies can lead to increased concentrations of chemical substances in the aquatic environment. To mitigate adverse effects on ecosystems and human health, the European Union established the **Water Framework Directive (Directive 2000/60/EC)**, which provides the foundation for integrated water resource management and pollution reduction at the source. Within this regulatory framework, **Environmental Quality Standards (EQS)** were introduced, as defined in **Directives 2008/105/EC and 2013/39/EU**.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In the immediate vicinity of a discharge point, concentrations of certain priority substances may temporarily exceed the established standards. The area surrounding the discharge point where such exceedances are permitted is referred to as the **mixing zone**.


>### I.3.	Scope of the Study
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Delineation of the mixing zone is a key element in assessing the environmental impact of wastewater discharges. It allows the spatial variation of pollutant concentrations after wastewater discharge to be analyzed and helps assess the risk of not meeting the environmental objectives established for the receiving water body.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The mechanisms governing the mixing of wastewater discharges in coastal waters are not uniform throughout the process and vary with distance from the discharge source. The physical processes controlling this behaviour are discussed in detail in ***Section II (Theoretical Foundations)***.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The main focus of this study is on the processes occurring in the immediate vicinity of the discharge point.The analysis of the near-field is crucial, as it defines the initial conditions for the subsequent transport and dispersion of pollutants in the marine environment. A detailed investigation of these processes provides the basis for the further development of tools that can improve the prediction of potential pollutant concentrations in marine waters, support the optimization of the depth and location of discharge pipelines, and provide a scientifically grounded assessment of whether a given project complies with regulatory water quality standards.


## II.	THEORETICAL FOUNDATIONS

>### II.1.	Discharge technologies  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Coastal outfalls are systems designed to convey and discharge treated wastewater from treatment facilities into a receiving water body. They can be classified based on their location (shoreline surface discharges or submerged outfalls), their mixing characteristics (single-port or multiport), and the properties of the discharged flow (positively or negatively buoyant). The primary function of the outfall structure is to ensure effective initial mixing of the discharge in the near-field region, which governs its subsequent transport and dispersion in the ambient environment.

>### II.2.	Physical description 
>>#### II.2.1.	Mixing Zones in Coastal Waters
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Most researchers divide the mixing process into two main regions based on the dominant physical mechanisms: the **near-field** and the **far-field**. Some authors identify a third phase, referred to as a transition region (*Koh & Brooks, 1975; Fischer et al., 1979*), while others consider it to be the final stage of the near-field (*Tsanis et al., 1994*).

>>>#### II.2.1.1.	Phase 1: Initial Mixing
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This phase begins immediately after the discharge of wastewater into the receiving water body. As the effluent  leaves the diffuser and enters the denser ambient seawater, the jet undergoes **initial dilution**. This stage is characterized by intense turbulent mixing, which leads to rapid dilution of the discharged effluent.The primary driving forces are the initial jet momentum (turbulence generated by the discharge velocity) and the buoyancy-induced flux.
>>>>

>>>#### I.3.2.	Phase 2: Plume Rise and Intermediate Mixing
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This phase occurs in the *mid-zone* and is characterized by dynamic horizontal dispersion and further vertical dilution. Differences in velocity and density between the effluent and the ambient water induce turbulence as the plume rises. The phase concludes when the plume reaches its terminal rise height. This height may correspond either to the water surface or to a specific depth where the density of the mixture becomes equal to that of the surrounding ambient water. In stratified waters, ambient density stratification reduces vertical mixing, limits the rise height of the jet, and may prevent it from reaching the surface.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The completion of this phase marks the end of the **near-field**, also referred to as the **hydrodynamic mixing zone** or the **initial dilution zone**.

>>>#### I.3.3.	Phase 3: Far-Field Transport
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The final phase of wastewater transport is characterized by the spreading of the discharged mixture under the influence of marine currents. The dynamic action of these currents causes dispersion of the mixture, leading to secondary mixing and dilution in the **far-field**. The rate of mixing is primarily controlled by the advection of marine currents and the turbulent characteristics of the ambient water, while the influence of the initial discharge parameters gradually diminishes. In certain cases, sedimentation of suspended particles on the seabed may also occur.

>>#### II.2.2.	Hydrodynamic Classification of Discharged Fluid Behaviour
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The behaviour of a discharged flow is primarily determined by the initial momentum of the jet, buoyancy forces arising from density differences, and the hydrodynamic characteristics of the receiving water body. These factors control the mixing processes and the subsequent evolution of the discharged fluid in the surrounding environment.  

>>>#### II.2.2.1.	Jet
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;An inertial, non-buoyant flow in which the behaviour of the discharge is dominated by **the initial momentum flux (M)** of the jet. In this regime, buoyancy forces are negligible, and the strong initial momentum generates intense turbulent mixing with the surrounding ambient water.

>>>#### II.2.2.2.	Plume
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A flow that resembles a jet in form, but whose development is primarily governed by **buoyancy forces** arising from density differences between the discharged fluid and the surrounding ambient water. Due to this density contrast, the flow may exhibit positive or negative buoyancy.

>>>#### II.2.2.3.	Buoyant Jet
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A flow in which the behaviour of the discharge is governed **by both the initial momentum of the jet and the buoyancy forces** arising from density differences between the discharged fluid and the surrounding ambient water. In the region close to the source, the flow is dominated by inertial forces characteristic of jets. As the distance from the source increases, the influence of the initial momentum gradually decreases and the flow begins to behave as a plume, where buoyancy becomes the dominant driving mechanism.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The relative importance of these two driving mechanisms is commonly described by the densimetric Froude number, which compares inertial forces to buoyancy forces. Depending on the value of this parameter, the discharge may behave closer to a momentum-dominated jet, a buoyancy-dominated plume, or an intermediate buoyant jet regime.


>### II.3.	Governing physical parameter
>>#### II.3.1.	Densimetric Froude number – Inertia vs Buoyancy
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The classical Froude number is a dimensionless parameter and represents the ratio of the specific kinetic energy of the flow to the specific potential energy, reflecting the energy structure of the flow and serving as a criterion for determining the flow regime (*Marinov et al., 2003*).  
$$Fr = \frac{\alpha v^2}{g h_{\text{avg}}} = [-] \, ,$$
where:  
$\alpha$ – Coriolis coefficient $[-]$  
$v$ – flow velocity $[m/s]$  
$g$ – gravitational acceleration $[m/s^2]$  
$h$ – mean flow depth $[m]$  

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Тhe flow regime according to the value of the Froude number is defined as follows:
>>* Fr < 1.0 – **subcritical flow** 
>>* Fr = 1.0 – **critical flow**  
>>* Fr > 1.0 – **supercritical flow**  

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;By introducing the expression:  
>>$$c = \sqrt{g h_{\text{avg}}} \, ,$$
>>which defines the propagation velocity of small elastic disturbances in the fluid, the Froude number can be represented as the ratio of inertial to gravitational forces (*Kazakov et al., 2008*):
>>$$Fr = \frac{v^2}{c^2} = \frac{\text{inertial forces}}{\text{gravitational forces}} [-]$$
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;When a density difference is present in the system, as is the case in mixing zone analysis where the difference occurs between the ambient fluid and the discharged jet, it is appropriate to express the classical Froude number as a ratio between inertial forces and gravitational (buoyancy) forces.
>>
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This ratio is known as the densimetric Froude number and is defined as:
>>$$Fr_d = \frac{v_p}{\sqrt{g' D_p}} [-] ,$$
>>where:  
>>$v_p$ – discharge velocity from the outlet/port/nozzle $[m/s]$  
>>$D_p$ – port/nozzle diameter $[m]$  
>>$g'$ – reduced gravitational acceleration $[m/s^2]$
>>$$g' = g \frac{\Delta \rho}{\rho_a} = g \frac{\rho_a - \rho_j}{\rho_a} [m/s^2] \, ,$$
>>where:  
>>$\rho_a$ – ambient fluid density $[kg/m^3]$  
>>$\rho_j$ – jet density $[kg/m^3]$
>>
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;According to the magnitude of the densimetric Froude number, the type of the jet can be interpreted as follows (*Hai-Bo Chen, 1991*):
>>
>>* $\text{Fr}_d \rightarrow \infty$ – **Jet** – the driving force is inertial; momentum is sufficiently large so that gravitational effects are negligible  
>>* $0 < \text{Fr}_d < \infty$ – **Buoyant Jet** – both inertial and buoyancy forces are present   
>>* $\text{Fr}_d \rightarrow 0$ – **Plume** – the driving force is gravitational; inertial effects are negligible


>>#### II.3.2.	Ambient stratification
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The description of the vertical stratification presented in this section refers specifically to the coastal waters of the Bulgarian Black Sea and is based on the analysis of climatic temperature and salinity fields reported in *Trukhchev & Naydenov (2018)*.
>>
>>>#### II.3.2.1.	Density stratification
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The vertical structure of the marine environment is characterized by density stratification, resulting from variations in temperature and salinity. In the Black Sea, stratification is particularly well developed and is primarily controlled by salinity, while temperature plays a significant role mainly in the upper layer.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The combined effect of these parameters leads to the formation of distinct vertical layers in the water column. In the surface layer, density variations are influenced by seasonal heating and freshwater input, whereas in deeper layers the density structure is predominantly governed by salinity. As a result, a stable stratified system is formed, which limits vertical exchange processes and strongly influences mixing in the marine environment.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;During the warm season, a secondary shallow thermocline develops in the upper layer of the water column, typically at depths between approximately 10.0 and 30.0 m, where significant vertical temperature gradients are observed. This layer can locally influence the initial mixing processes in the near-field region, particularly for shallow discharges.

>>>#### II.3.2.2.	Pycnocline
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A key feature of the vertical density structure is the presence of a well-defined pycnocline, typically located at depths between approximately 50.0 and 120.0 m. This layer is characterized by strong vertical gradients of density, resulting mainly from changes in salinity.  
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The pycnocline acts as a barrier to vertical mixing, separating the upper, relatively well-mixed layer from the deeper, more stable water masses. Due to the high density gradients within this layer, vertical transport of mass and momentum is significantly reduced, which has a direct impact on the development of buoyant jets and plumes in the near-field region.


>### II.4.	Mathematical Representation of the Flow
>>#### II.4.1.	Development of Buoyant Jet and Plume Theory
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The study of turbulent buoyant jets and plumes has developed significantly since the early 20th century, beginning with experimental investigations by *Rawn and Palmer (1930)*. A major advancement was introduced by *Morton, Taylor and Turner (1956)*, who formulated the integral approach for buoyant plumes and introduced the concept of entrainment. Subsequent studies extended this framework to account for various discharge conditions and ambient environments (*Abraham, 1963; Fan, 1967*). A comprehensive synthesis of these developments was later presented by *Fischer et al. (1979)*, which remains a fundamental reference for the analysis of mixing processes in inland and coastal waters. More recent developments have focused on engineering applications and predictive modeling of mixing zones and marine outfalls, particularly through the work of *Jirka, Roberts, and Lee & Cheung*.
These modeling approaches are fundamentally derived from the governing equations of fluid motion, originating from the *Euler equations* for ideal fluids and later extended to real fluids through the *Navier–Stokes equations*, which form the theoretical basis for the description of turbulent buoyant jets and plumes.

>>#### II.4.2.	Navier–Stokes Equations
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The mathematical description of any fundamental problem in fluid dynamics is rooted in the *Navier–Stokes equations*. These equations treat the fluid as a continuum and represent the fundamental laws of conservation of **mass**, **momentum**, and **energy**.
>>
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In their most general form, these governing equations are expressed as follows:
>>
>>**1)	Conservation of mass:**
$$\frac{\partial \rho}{\partial t} + \frac{\partial (\rho u_i)}{\partial x_i} = \frac{\partial \rho}{\partial t} + \rho \frac{\partial u_i}{\partial x_i} + u_i \frac{\partial \rho}{\partial x_i} = 0$$

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The individual terms in the mass equation represent the effects of local density change with time and advective mass transfer.

>>**2)	Conservation of momentum:**
$$\frac{\partial (\rho u_i)}{\partial t} + \frac{\partial (u_j \rho u_i)}{\partial x_j} - \frac{\partial \tau_{ij}}{\partial x_j} + \frac{\partial p}{\partial x_i} = 0$$

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The individual terms in the momentum equation represent the effects of local acceleration, advective acceleration, viscous forces, pressure gradient, and volume forces.

>>**3)	Conservation of energy:**
$$\frac{\partial (\rho \boldsymbol{E})}{\partial t} + \frac{\partial (u_j \rho \boldsymbol{E})}{\partial x_j} - \frac{\partial}{\partial x_i} \left( k \frac{\partial T}{\partial x_i} \right) + \frac{\partial (u_j p)}{\partial x_j} - \frac{\partial \tau_{ij} u_j}{\partial x_j} = 0$$

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The individual terms in the energy conservation equation represent the effects of local energy change with time, advective energy transfer, thermal conductivity, work done by pressure, work done by viscous forces.

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The mathematical description of hydrodynamic processes during wastewater discharge is based on the *Navier–Stokes equation*s. Due to their nonlinear nature, direct analytical solutions of these equations are not feasible, while full-scale numerical simulations (CFD) of turbulent flows at large scales are computationally expensive and require significant resources.Therefore, for the purposes of applied modeling, a set of simplifying assumptions is introduced. These enable the formulation of a more efficient computational model that preserves physical fidelity while significantly reducing complexity.

>>#### II.4.3.	Entrainment Hypothesis
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A key mechanism governing the evolution of turbulent jets and plumes is the entrainment of ambient fluid into the flow. This process represents the incorporation of surrounding fluid into the jet due to turbulent mixing at its boundary, leading to an increase in volume flux and a gradual dilution of the plume.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The concept of entrainment was first introduced by *Stommel (1947)* and later formalized in the classical work of *Morton, Taylor and Turner (1956)*, who proposed that the entrainment velocity is proportional to a characteristic velocity of the flow (*Matulka et al., 2014*).
$$u_e = \alpha u \, ,$$
where:  
$u_e$ – entrainment velocity $[m/s]$  
$u$ – characteristic velocity $[m/s]$  
$\alpha$ – entrainment coefficient $[-]$  



## III.	MODEL FORMULATION

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In this study, a simplified modeling approach is adopted based on an integral formulation of the governing equations. The adopted integral approach follows the classical formulation introduced by *Morton, Taylor and Turner (1956)* and further developed in environmental applications (*Fischer et al., 197*). The integral modeling are based on the conservation of **mass**, **momentum**, and **buoyancy** and represents a well-established and widely applied method for the analysis of buoyant jets in near-field conditions. It is considered a standard engineering tool due to its ability to accurately predict key characteristics such as plume trajectory and initial dilution, while maintaining low computational requirements.  
The main advantage of this approach lies in its efficiency and robustness, making it particularly suitable for practical applications, including environmental impact assessments of submarine outfalls.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The primary trade-off associated with the use of an integral model is the loss of detailed representation of the turbulent flow structure (*Fischer et al., 1979; Pope, 2000*). While the full *Navier–Stokes equations*, solved using high-fidelity approaches such as *Direct Numerical Simulation (DNS)* or *Large Eddy Simulation (LES)*, are capable of resolving the full spectrum of turbulent scales, the integral approach operates on averaged flow quantities.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;As a result, the model does not explicitly:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;➢Instantaneous fluctuations of velocity and scalar concentration, as only mean flow properties are considered  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;➢Anisotropy of turbulence within the mixing region  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;➢Fine-scale turbulent structures at the plume boundary responsible for detailed mixing processes  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Despite these limitations, the integral model remains highly suitable for engineering applications. In the context of environmental assessment, the primary quantities of interest are the mean dilution and the terminal rise height of the plume, rather than the detailed turbulent structure.
