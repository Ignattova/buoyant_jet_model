## I. INTRODUCTION

>### I.1 Problem Statement and Motivation
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Coastal regions hold significant environmental and economic value at both global and national scales. The demand for recreation and tourism has led to a high concentration of population in these areas, resulting in rapid urban expansion and increasing pressure on coastal zones. One of the primary sources of this environmental pressure is the discharge of municipal and industrial wastewater.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;*The Bulgarian Black Sea coast* is the main recipient of a significant part of the wastewater in the country - wastewater from urban treatment plants (WWTPs), sewage without complete treatment, as well as significant volumes of urban stormwater runoff. This leads to increased anthropogenic pollution and biogenic nutrient enrichment (eutrophication), which has become a major ecological concern over recent decades. As a result, the entire Bulgarian Black Sea coast has been designated as a ***sensitive area***.


>### I.2 Environmental Quality Standards and Mixing Zones
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The discharge of wastewater of different composition into water bodies can lead to increased concentrations of chemical substances in the aquatic environment. To mitigate adverse effects on ecosystems and human health, the European Union established the **Water Framework Directive (Directive 2000/60/EC)**, which provides the foundation for integrated water resource management and pollution reduction at the source. Within this regulatory framework, **Environmental Quality Standards (EQS)** were introduced, as defined in **Directives 2008/105/EC and 2013/39/EU**.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In the immediate vicinity of a discharge point, concentrations of certain priority substances may temporarily exceed the established standards. The area surrounding the discharge point where such exceedances are permitted is referred to as the **mixing zone**.


>### I.3.	Scope of the Study
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Delineation of the mixing zone is a key element in assessing the environmental impact of wastewater discharges. It allows the spatial variation of pollutant concentrations after wastewater discharge to be analyzed and helps assess the risk of not meeting the environmental objectives established for the receiving water body.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The mechanisms governing the mixing of wastewater discharges in coastal waters are not uniform throughout the process and vary with distance from the discharge source. The physical processes controlling this behaviour are discussed in detail in ***Section II (Theoretical Foundations)***.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The main focus of this study is on the processes occurring in the immediate vicinity of the discharge point.The analysis of the near-field is crucial, as it defines the initial conditions for the subsequent transport and dispersion of pollutants in the marine environment. A detailed investigation of these processes provides the basis for the further development of tools that can improve the prediction of potential pollutant concentrations in marine waters, support the optimization of the depth and location of discharge pipelines, and provide a scientifically grounded assessment of whether a given project complies with regulatory water quality standards.


## II.	THEORETICAL FOUNDATIONS

>### II.1.	Discharge technologies  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Coastal outfalls are systems designed to convey and discharge treated wastewater from treatment facilities into a receiving water body. They can be classified based on their location (shoreline surface discharges or submerged outfalls), their mixing characteristics (single-port or multiport), and the properties of the discharged flow (positively or negatively buoyant). The primary function of the outfall structure is to ensure effective initial mixing of the discharge in the near-field region, which governs its subsequent transport and dispersion in the ambient environment.

>### II.2.	Physical description 
>>#### II.2.1.	Mixing Zones in Coastal Waters
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Most researchers divide the mixing process into two main regions based on the dominant physical mechanisms: the **near-field** and the **far-field**. Some authors identify a third phase, referred to as a transition region (*Koh & Brooks, 1975; Fischer et al., 1979*), while others consider it to be the final stage of the near-field (*Tsanis et al., 1994*).

>>>#### II.2.1.1.	Phase 1: Initial Mixing
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This phase begins immediately after the discharge of wastewater into the receiving water body. As the effluent  leaves the diffuser and enters the denser ambient seawater, the jet undergoes **initial dilution**. This stage is characterized by intense turbulent mixing, which leads to rapid dilution of the discharged effluent.The primary driving forces are the initial jet momentum (turbulence generated by the discharge velocity) and the buoyancy-induced flux.
>>>>

>>>#### II.2.1.2.	Phase 2: Plume Rise and Intermediate Mixing
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This phase occurs in the *mid-zone* and is characterized by dynamic horizontal dispersion and further vertical dilution. Differences in velocity and density between the effluent and the ambient water induce turbulence as the plume rises. The phase concludes when the plume reaches its terminal rise height. This height may correspond either to the water surface or to a specific depth where the density of the mixture becomes equal to that of the surrounding ambient water. In stratified waters, ambient density stratification reduces vertical mixing, limits the rise height of the jet, and may prevent it from reaching the surface.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The completion of this phase marks the end of the **near-field**, also referred to as the **hydrodynamic mixing zone** or the **initial dilution zone**.

>>>#### II.2.1.3.	Phase 3: Far-Field Transport
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The final phase of wastewater transport is characterized by the spreading of the discharged mixture under the influence of marine currents. The dynamic action of these currents causes dispersion of the mixture, leading to secondary mixing and dilution in the **far-field**. The rate of mixing is primarily controlled by the advection of marine currents and the turbulent characteristics of the ambient water, while the influence of the initial discharge parameters gradually diminishes. In certain cases, sedimentation of suspended particles on the seabed may also occur.

>>#### II.2.2.	Hydrodynamic Classification of Discharged Fluid Behaviour
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The behaviour of a discharged flow is primarily determined by the initial momentum of the jet, buoyancy forces arising from density differences, and the hydrodynamic characteristics of the receiving water body. These factors control the mixing processes and the subsequent evolution of the discharged fluid in the surrounding environment.  

>>>#### II.2.2.1.	Jet
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;An inertial, non-buoyant flow in which the behaviour of the discharge is dominated by **the initial momentum flux (M)** of the jet. In this regime, buoyancy forces are negligible, and the strong initial momentum generates intense turbulent mixing with the surrounding ambient water.

>>>#### II.2.2.2.	Plume
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A flow that resembles a jet in form, but whose development is primarily governed by **buoyancy forces** arising from density differences between the discharged fluid and the surrounding ambient water. Due to this density contrast, the flow may exhibit positive or negative buoyancy.

>>>#### II.2.2.3.	Buoyant Jet
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A flow in which the behaviour of the discharge is governed **by both the initial momentum of the jet and the buoyancy forces** arising from density differences between the discharged fluid and the surrounding ambient water. In the region close to the source, the flow is dominated by inertial forces characteristic of jets. As the distance from the source increases, the influence of the initial momentum gradually decreases and the flow begins to behave as a plume, where buoyancy becomes the dominant driving mechanism.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The relative importance of these two driving mechanisms is commonly described by the densimetric Froude number, which compares inertial forces to buoyancy forces. Depending on the value of this parameter, the discharge may behave closer to a momentum-dominated jet, a buoyancy-dominated plume, or an intermediate buoyant jet regime.


>### II.3.	Governing physical parameter
>>#### II.3.1.	Densimetric Froude number – Inertia vs Buoyancy
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The classical Froude number is a dimensionless parameter and represents the ratio of the specific kinetic energy of the flow to the specific potential energy, reflecting the energy structure of the flow and serving as a criterion for determining the flow regime (*Marinov et al., 2003*).  
$$Fr = \frac{\alpha v^2}{g h_{\text{avg}}} = [-] \, ,$$
where:  
$\alpha$ – Coriolis coefficient $[-]$  
$v$ – flow velocity $[m/s]$  
$g$ – gravitational acceleration $[m/s^2]$  
$h$ – mean flow depth $[m]$  

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Тhe flow regime according to the value of the Froude number is defined as follows:  
>>➢ $\text{Fr}$ < 1.0 – **subcritical flow**  
>>➢ $\text{Fr}$ = 1.0 – **critical flow**   
>>➢ $\text{Fr}$ > 1.0 – **supercritical flow**  

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;By introducing the expression:  
>>$$c = \sqrt{g h_{\text{avg}}} \, ,$$
>>which defines the propagation velocity of small elastic disturbances in the fluid, the Froude number can be represented as the ratio of inertial to gravitational forces (*Kazakov et al., 2008*):
>>$$Fr = \frac{v^2}{c^2} = \frac{\text{inertial forces}}{\text{gravitational forces}} [-]$$
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;When a density difference is present in the system, as is the case in mixing zone analysis where the difference occurs between the ambient fluid and the discharged jet, it is appropriate to express the classical Froude number as a ratio between inertial forces and gravitational (buoyancy) forces.
>>
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This ratio is known as the densimetric Froude number and is defined as:
>>$$Fr_d = \frac{v_p}{\sqrt{g' D_p}} [-] ,$$
>>where:  
>>$v_p$ – discharge velocity from the outlet/port/nozzle $[m/s]$  
>>$D_p$ – port/nozzle diameter $[m]$  
>>$g'$ – reduced gravitational acceleration $[m/s^2]$
>>$$g' = g \frac{\Delta \rho}{\rho_a} = g \frac{\rho_a - \rho_j}{\rho_a} [m/s^2] \, ,$$
>>where:  
>>$\rho_a$ – ambient fluid density $[kg/m^3]$  
>>$\rho_j$ – jet density $[kg/m^3]$
>>
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;According to the magnitude of the densimetric Froude number, the type of the jet can be interpreted as follows (*Hai-Bo Chen, 1991*):
>>
>>➢ $\text{Fr}_d \rightarrow \infty$ – **Jet** – the driving force is inertial; momentum is sufficiently large so that gravitational effects are negligible  
>>➢ $0 < \text{Fr}_d < \infty$ – **Buoyant Jet** – both inertial and buoyancy forces are present   
>>➢ $\text{Fr}_d \rightarrow 0$ – **Plume** – the driving force is gravitational; inertial effects are negligible


>>#### II.3.2.	Ambient stratification
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The description of the vertical stratification presented in this section refers specifically to the coastal waters of the Bulgarian Black Sea and is based on the analysis of climatic temperature and salinity fields described in the report by *Trukhchev & Naydenov (2018)*.
>>
>>>#### II.3.2.1.	Density stratification
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The vertical structure of the marine environment is characterized by density stratification, resulting from variations in temperature and salinity. In the Black Sea, stratification is particularly well developed and is primarily controlled by salinity, while temperature plays a significant role mainly in the upper layer.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The combined effect of these parameters leads to the formation of distinct vertical layers in the water column. In the surface layer, density variations are influenced by seasonal heating and freshwater input, whereas in deeper layers the density structure is predominantly governed by salinity. As a result, a stable stratified system is formed, which limits vertical exchange processes and strongly influences mixing in the marine environment.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;During the warm season, a secondary shallow thermocline develops in the upper layer of the water column, typically at depths between approximately 10.0 and 30.0 m, where significant vertical temperature gradients are observed. This layer can locally influence the initial mixing processes in the near-field region, particularly for shallow discharges.

>>>#### II.3.2.2.	Pycnocline
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A key feature of the vertical density structure is the presence of a well-defined pycnocline, typically located at depths between approximately 50.0 and 120.0 m. This layer is characterized by strong vertical gradients of density, resulting mainly from changes in salinity.  
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The pycnocline acts as a barrier to vertical mixing, separating the upper, relatively well-mixed layer from the deeper, more stable water masses. Due to the high density gradients within this layer, vertical transport of mass and momentum is significantly reduced, which has a direct impact on the development of buoyant jets and plumes in the near-field region.


>### II.4.	Mathematical Representation of the Flow
>>#### II.4.1.	Development of Buoyant Jet and Plume Theory
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The study of turbulent buoyant jets and plumes has developed significantly since the early 20th century, beginning with experimental investigations by *Rawn and Palmer (1930)*. A major advancement was introduced by *Morton, Taylor and Turner (1956)*, who formulated the integral approach for buoyant plumes and introduced the concept of entrainment. Subsequent studies extended this framework to account for various discharge conditions and ambient environments (*Abraham, 1963; Fan, 1967*). A comprehensive synthesis of these developments was later presented by *Fischer et al. (1979)*, which remains a fundamental reference for the analysis of mixing processes in inland and coastal waters. More recent developments have focused on engineering applications and predictive modeling of mixing zones and marine outfalls, particularly through the work of *Jirka, Roberts, and Lee & Cheung*.
These modeling approaches are fundamentally derived from the governing equations of fluid motion, originating from the *Euler equations* for ideal fluids and later extended to real fluids through the *Navier–Stokes equations*, which form the theoretical basis for the description of turbulent buoyant jets and plumes.

>>#### II.4.2.	Navier–Stokes Equations
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The mathematical description of any fundamental problem in fluid dynamics is rooted in the *Navier–Stokes equations*. These equations treat the fluid as a continuum and represent the fundamental laws of conservation of **mass**, **momentum**, and **energy**.
>>
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In their most general form, these governing equations are expressed as follows:
>>
>>>**1)	Conservation of mass:**
$$\frac{\partial \rho}{\partial t} + \frac{\partial (\rho v_i)}{\partial x_i} = \frac{\partial \rho}{\partial t} + \rho \frac{\partial v_i}{\partial x_i} + v_i \frac{\partial \rho}{\partial x_i} = 0$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The individual terms in the mass equation represent the effects of local density change with time and advective mass transfer.

>>>**2)	Conservation of momentum:**
$$\frac{\partial (\rho v_i)}{\partial t} + \frac{\partial (v_j \rho v_i)}{\partial x_j} - \frac{\partial \tau_{ij}}{\partial x_j} + \frac{\partial p}{\partial x_i} = 0$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The individual terms in the momentum equation represent the effects of local acceleration, advective acceleration, viscous forces, pressure gradient, and volume forces.

>>>*3)	Conservation of energy:**
$$\frac{\partial (\rho \boldsymbol{E})}{\partial t} + \frac{\partial (v_j \rho \boldsymbol{E})}{\partial x_j} - \frac{\partial}{\partial x_i} \left( k \frac{\partial T}{\partial x_i} \right) + \frac{\partial (v_j p)}{\partial x_j} - \frac{\partial \tau_{ij} v_j}{\partial x_j} = 0$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The individual terms in the energy conservation equation represent the effects of local energy change with time, advective energy transfer, thermal conductivity, work done by pressure, work done by viscous forces.

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The mathematical description of hydrodynamic processes during wastewater discharge is based on the *Navier–Stokes equation*s. Due to their nonlinear nature, direct analytical solutions of these equations are not feasible, while full-scale numerical simulations (CFD) of turbulent flows at large scales are computationally expensive and require significant resources.Therefore, for the purposes of applied modeling, a set of simplifying assumptions is introduced. These enable the formulation of a more efficient computational model that preserves physical fidelity while significantly reducing complexity.

>>#### II.4.3.	Entrainment Hypothesis
>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A key mechanism governing the evolution of turbulent jets and plumes is the entrainment of ambient fluid into the flow. This process represents the incorporation of surrounding fluid into the jet due to turbulent mixing at its boundary, leading to an increase in volume flux and a gradual dilution of the plume.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The concept of entrainment was first introduced by *Stommel (1947)* and later formalized in the classical work of *Morton, Taylor and Turner (1956)*, who proposed that the entrainment velocity is proportional to a characteristic velocity of the flow (*Matulka et al., 2014*).
$$v_e = \alpha v \, ,$$
where:  
$v_e$ – entrainment velocity $[m/s]$  
$v$ – characteristic velocity $[m/s]$  
$\alpha$ – entrainment coefficient $[-]$  



## III.	MODEL FORMULATION

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In this study, a simplified modeling approach is adopted based on an integral formulation of the governing equations. The adopted integral approach follows the classical formulation introduced by *Morton, Taylor and Turner (1956)* and further developed in environmental applications (*Fischer et al., 1979*). The integral modeling are based on the conservation of **mass**, **momentum**, and **buoyancy** and represents a well-established and widely applied method for the analysis of buoyant jets in near-field conditions. It is considered a standard engineering tool due to its ability to accurately predict key characteristics such as plume trajectory and initial dilution, while maintaining low computational requirements.  
The main advantage of this approach lies in its efficiency and robustness, making it particularly suitable for practical applications, including environmental impact assessments of submarine outfalls.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The primary trade-off associated with the use of an integral model is the loss of detailed representation of the turbulent flow structure (*Fischer et al., 1979; Pope, 2000*). While the full *Navier–Stokes equations*, solved using high-fidelity approaches such as *Direct Numerical Simulation (DNS)* or *Large Eddy Simulation (LES)*, are capable of resolving the full spectrum of turbulent scales, the integral approach operates on averaged flow quantities.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;As a result, the model does not explicitly:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;➢Instantaneous fluctuations of velocity and scalar concentration, as only mean flow properties are considered  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;➢Anisotropy of turbulence within the mixing region  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;➢Fine-scale turbulent structures at the plume boundary responsible for detailed mixing processes  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Despite these limitations, the integral model remains highly suitable for engineering applications. In the context of environmental assessment, the primary quantities of interest are the mean dilution and the terminal rise height of the plume, rather than the detailed turbulent structure.  




>### III.1.	Assumptions
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;To formulate a mathematical model, the following assumptions are introduced:  
>> **1)	Steady-state process**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The flow is assumed to be steady, meaning that the characteristics of the source and the ambient environment do not vary in time.  
**2)	Single representative jet**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The ports of the multiport diffuser are considered to be hydrodynamically independent. When the spacing between them is sufficiently large, the interaction between individual jets can be neglected, and a single representative jet is considered.  
**3)	Near-field model**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The model is applicable only to the near-field region, where the jet behavior is primarily governed by the interaction between momentum, buoyancy, and turbulent entrainment. Far-field processes are not considered  
**4)	Quiescent ambient (no currents)**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The ambient environment is assumed to be at rest, and the influence of external currents is neglected.  
**5)	Axisymmetric flow**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The jet is assumed to be symmetrical about its central axis.  
**6)	Turbulent flow**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The flow is turbulent with a high Reynolds number. Molecular viscosity is neglected, and turbulent mixing is represented through the entrainment hypothesis.  
**7)	Boussinesq approximation**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Density differences between the jet and the ambient fluid are assumed to be small and are considered only in the buoyancy term, without affecting the inertial term.  
$$\Delta \rho \ll \rho_0$$
$$\rho \approx \rho_0$$
**8)	Stratified ambient (two-layer model)**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The ambient environment is represented as a two-layer system with constant density in each layer   
**9)	Hydrostatic pressure**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The pressure field is assumed to be hydrostatic, and dynamic variations in the vertical direction are neglected.  
**10)	Neglect of longitudinal diffusion**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Variations along the jet axis are assumed to be much smaller than variations across the jet cross-section, which allows the neglect of longitudinal diffusion.  
**11)	Cross-sectional averaging (top-hat profile)**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Velocity and density are represented by averaged values over a given cross-section, assuming a uniform distribution across the jet.  
**12)	Entrainment hypothesis**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The entrainment of ambient fluid into the jet is assumed to be proportional to the characteristic velocity.  

>### III.2.	Integral model
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The integral approach is based on the transformation of the fundamental conservation laws (of **mass**, **momentum** and **energy/buoyancy**) from their differential form (*Navier-Stokes equations*) into integral form. Following integral method and the adopted cross-sectional averaging of the jet, the flow is described in terms of three integral quantities: **volume flux**, **momentum flux**, and **buoyancy flux** (*Fischer et al., 1979*). In accordance with the Boussinesq approximation, density is assumed to be constant in the conservation equations for mass and momentum, while the effect of density variations is accounted for only in the buoyancy term through the reduced gravity $g'$.

>> **1)	The volume flux $[m^3/s]$ is defined as:**
>$$Q = \int_{A} v \, dA$$
>where $v$ is the axial velocity and $A$ is the jet cross-sectional area.
>> 
>> **2)The momentum flux $[m^4/s^2]$ is defined as:**
>$$M = \int_{A} v^2 \, dA$$
>and represents the transport of momentum through a cross-section of area $A$.
>> 
>> **3)The buoyancy flux $[m^4/s^3]$ is defined as:**
>$$B = \int_{A} g' v \, dA$$
>where:  
>>  $g'$ - reduced gravity associated with the density difference between the jet and the ambient fluid $[m/s^2]$
>> 
>>and represents the effect of the density difference between the discharged fluid and the ambient environment under the action of gravity. It quantifies the tendency of the jet to rise (*positive buoyancy*) or to sink (*negative buoyancy*).

>### III.3.	Ordinary Differential Equations system
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Once the integral quantities **Q**, **M**, and **B** have been defined, the next step is to apply the conservation principles of **mass**, **momentum**, and **buoyancy** to a control volume moving along the jet. Under the assumptions adopted in the present study the governing equations can be expressed in a one-dimensional integral form. In this way, the *original partial differential equations* are represented by a system of *ordinary differential equations* describing the change of the jet along its trajectory, including the increase of volume flux due to entrainment of ambient fluid, the variation of momentum under buoyancy effects, and the change in the trajectory geometry (angle and position).  

>>#### III.3.1.	Coordinate system and model setup
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The flow is described in a two-dimensional vertical plane using a **Cartesian coordinate system**. The horizontal coordinate $x$ represents the along-flow direction, while the vertical coordinate $z$ measures depth relative to the seabed. The total water depth is denoted by $H$, such that the free surface is located at $z = H$.
The discharge point is located at $(x, z) = (0, z_0)$, where $z_0$  is the elevation of the outlet above the seabed.
The jet trajectory is parameterized by the **arc-length coordinate** $s$, measured along the centerline of the flow. The local jet direction is described by the **angle** $\theta$, defined with respect to the horizontal axis.
The ambient environment is represented by a **two-layer stratified system**, with a density interface (pycnocline) located at elevation $z_{pyc}$. The lower layer is characterized by density $\rho_{bottom}$, while the upper layer has density $\rho_{top}$  

>>#### III.3.2. Governing ODE system

>>>#### III.3.2.1.	Volume Flux Conservation Equation  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The variation of the volume flux along the jet trajectory is governed by the entrainment of ambient fluid into the jet. Under the top hat assumption and for an axisymmetric jet, the conservation equation for the volume flux can be written as:

>>>$$\frac{dQ}{ds} = 2\alpha \sqrt{\pi M}$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;where α is the entrainment coefficient.

>>>#### III.3.2.2.	Momentum Flux Conservation Equation 
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In a quiescent ambient, the only force capable of altering the jet momentum is the buoyancy force. Since it acts in the vertical direction, it affects only the vertical component of the momentum, which leads to a change in the jet trajectory. Using the adopted definitions of the integral quantities, the momentum equation is written as:

>>>$$\frac{dM}{ds} = \frac{BQ}{M} \sin \theta$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This equation shows how the momentum flux varies under the action of buoyancy as the jet changes direction in the receiving water.

>>>#### III.3.2.3.	Buoyancy Flux Conservation Equation
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Within each homogeneous layer, the ambient density is assumed constant. As a result, the reduced gravity g′ is also constant and the buoyancy flux is conserved along the jet trajectory:

>>>$$\frac{dB}{ds} = 0$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In the present study, however, the receiving water is represented by a two layer stratified system. Therefore, the buoyancy flux remains constant only within each layer and is redefined when the jet crosses the density interface (pycnocline).  
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The ambient density is defined as:

>>>$$\rho_a(z) = \begin{cases} \rho_{bottom'}, & z < z_{pyc} \\ \rho_{top'}, & z \geq z_{pyc} \end{cases}$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;with the pycnocline being considered part of the upper layer.
>>> 
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Reduced gravity is correspondingly written as:

>>>$$g'(z) = g \frac{\rho_a(z) - \rho_0}{\rho_a(z)}$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Thus, the buoyancy flux can be expressed as:

>>>$$B(z) = g'(z) Q(z)$$
>>>
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This formulation allows the effect of stratification to be included while keeping the model within the adopted integral framework.

>>>#### III.3.2.4.	Geometrical equations
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The change in the jet trajectory is described by the following geometric relationships:  

>>>$$\frac{dx}{ds} = \cos \theta$$

>>>$$\frac{dz}{ds} = \sin \theta$$

>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;where $x$ and $z$ are the horizontal and vertical coordinates of the jet centerline, respectively, and $\theta$ is the angle describing the deflection of the centerline relative to the horizontal axis.
>>>
>>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;These equations define the jet trajectory in the vertical direction.

>>#### III.3.3. Input Parameters

>>>#### III.3.3.1. Discharge parameters  
➢ $D$ — nozzle/port/pipe diameter $[m]$  
➢ $V_0$ — initial discharge velocity $[m/s]$  
➢ $\theta_0$ — initial discharge angle $[^\circ]$   
➢$\rho_0$ — density of the discharged fluid $[kg/m^3]$  

>>>#### III.3.3.2. Ambient parameters  
➢ $H$ — total water depth $[m]$  
➢ $z_0$ — elevation of the outlet above the seabed $[m]$  
➢ $z_{pyc}$ — elevation of the pycnocline above the seabed $[m]$  
➢ $\rho_{bottom}$ — density of the lower layer $[kg/m^3]$  
➢ $\rho_{top}$ — density of the upper layer $[kg/m^3]$  

>>>#### III.3.3.3. Model parameter  
➢ $\alpha$ — entrainment coefficient $[-]$  

>>#### III.3.4. Initial Conditions
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The initial conditions are defined at the outlet of the discharge pipe. At $s = 0$ (pipe outlet), the initial values of the integral quantities are obtained directly from their integral definitions under the assumption of a uniform (**top-hat**) velocity distribution, where the velocity is constant across the cross-section, $V = const$. As a result, the initial **volume flux**, **momentum flux**, and **buoyancy flux** are expressed as follows:

$$Q_0 = V_0 \frac{\pi D^2}{4}$$
$$M_0 = V_0 Q_0$$
$$B_0 = g'_0 Q_0$$

>>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The initial position and direction of the jet are taken as:
$$x(0) = 0, \quad z(0) = z_0, \quad \theta(0) = \theta_0$$

>>#### III.3.5. Simulation Cases
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;To investigate the behavior of the jet under different hydrodynamic conditions, several computational cases were considered. The main objective is to analyze the influence of ambient stratification and the initial discharge parameters on the jet trajectory and the dilution process.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A base case is defined as a reference point for comparison, after which variants with modified parameters are considered. In each case, only one factor is varied, while the remaining parameters are kept constant in order to clearly distinguish its influence.


>>>#### III.3.5.1.	Base Case  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The ambient environment is represented as a two-layer system separated by a seasonal thermocline. This assumption reflects typical summer conditions in the shallow coastal zone, where intense solar heating creates a pronounced temperature boundary. The lower layer is assumed to consist of colder and denser water, while the upper layer is represented by a well-mixed and warmed surface layer with lower density. The ambient and discharge parameters are introduced as representative average values consistent with realistic field conditions.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;For the base case, a horizontal discharge is assumed ($\theta_0 = 0$). Under this initial configuration, the jet is expected to follow a longer centerline path within the water column before reaching the free surface, which may promote greater entrainment of ambient water and, consequently, greater initial dilution. This case is used as the reference scenario.  

>>> ➢ $D$ = 0.5552 $[m]$  
➢ $V_0$ = 1.0 $[m/s]$  
➢ $\theta_0$ = 0 $[^\circ]$   
➢ $\rho_0$ = 998.5 $[kg/m^3]$  
➢ $H$ = 16.0 $[m]$  
➢ $z_0$ = 0.5 $[m]$  
➢ $z_{pyc}$ = 6.0 $[m]$  
➢ $\rho_{bottom}$ = 1014.5 $[kg/m^3]$  
➢ $\rho_{top}$ = 1011.0 $[kg/m^3]$  
➢ $\alpha$ = 0.09 $[-]$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Two options for changing the entrainment coefficient will be considered:
  
>>>➢ $\alpha$ = 0.07 $[-]$  
➢ $\alpha$ = 0.13 $[-]$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Two options for changing the discharge angle will be considered:

>>#### III.3.6. Numerical Implementation 

>>>#### III.3.6.1.	Solver Selection
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The system of ordinary differential equations is solved numerically using the ***solve_ivp function (RK45 method)*** from the ***SciPy library***. Due to the presence of a stratified environment, the integration is performed in two consecutive stages — one for the lower layer and one for the upper layer. Within each homogeneous layer, the buoyancy flux $B$ is assumed to be constant, and it is recalculated when the jet crosses the pycnocline. ***Event functions*** are employed to control the solution and to detect when the jet reaches the free surface, the seabed, or the interface between the two layers.

>>>#### III.3.6.2.	Code Structure: 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from dataclasses import dataclass


# =========================================================
# 1. PARAMETERS OF THE OUTFALL / AMBIENT
# =========================================================
@dataclass
class OutfallParams:
    D: float = 0.5552             # Internal pipe diameter [m]
    rho_0: float = 998.5          # Effluent density [kg/m3]
    z_0: float = 0.5              # Outlet elevation above seabed [m]
    H_total: float = 16.0         # Total water depth [m]
    z_pyc: float = 6.0            # Pycnocline elevation above seabed [m]
    rho_bottom: float = 1014.5    # Density below the pycnocline [kg/m3]
    rho_top: float = 1011.0       # Density above the pycnocline [kg/m3]
    g: float = 9.81               # Gravity [m/s2]


# =========================================================
# 2. ODE SYSTEM FOR A SINGLE HOMOGENEOUS LAYER
# =========================================================
def jet_equations(s, state, alpha, B_const):
    """
    Differential equations for one homogeneous layer.

    state = [Q, M, theta, x, z]

    Q     - volume flux
    M     - momentum flux 
    theta - trajectory angle [rad]
    x     - horizontal coordinate [m]
    z     - vertical coordinate above seabed [m]
    """
    Q, M, theta, x, z = state

    # Protection against division by zero
    Q_safe = max(Q, 1e-12)
    M_safe = max(M, 1e-12)

    # Jet radius
    r_effective = np.sqrt(Q_safe**2 / (np.pi * M_safe))
    V_char = M_safe / Q_safe

    # 1) Entrainment equation
    dQ_ds = 2.0 * np.pi * alpha * r_effective * V_char

    # 2) Momentum equation
    dM_ds = (B_const * Q_safe / M_safe) * np.sin(theta)

    # 3) Change of trajectory angle
    dtheta_ds = (B_const * Q_safe / M_safe**2) * np.cos(theta)

    # 4) Geometry of the centerline
    dx_ds = np.cos(theta)
    dz_ds = np.sin(theta)

    return [dQ_ds, dM_ds, dtheta_ds, dx_ds, dz_ds]


# =========================================================
# 3. HELPER FUNCTIONS
# =========================================================
def calculate_initial_conditions(V0, theta0_deg, params):
    """Calculate Q0, M0 and theta0 at the source."""
    Q0 = (np.pi * params.D**2 / 4.0) * V0
    M0 = Q0 * V0
    theta0 = np.deg2rad(theta0_deg)
    return Q0, M0, theta0


def calculate_reduced_gravity(rho_ambient, rho_0, g):
    """Reduced gravity g'."""
    return g * (rho_ambient - rho_0) / rho_ambient


def create_events(params):
    """Create event functions for pycnocline, free surface and bottom."""

    def hit_pycnocline(s, state):
        z = state[4]
        return z - params.z_pyc
    hit_pycnocline.terminal = True
    hit_pycnocline.direction = 1

    def hit_surface(s, state):
        z = state[4]
        return params.H_total - z
    hit_surface.terminal = True
    hit_surface.direction = -1

    def hit_bottom(s, state):
        z = state[4]
        return z
    hit_bottom.terminal = True
    hit_bottom.direction = -1

    return hit_pycnocline, hit_surface, hit_bottom


def solve_one_layer(y0, alpha, B_const, s_start, s_end, events):
    """Solve the ODE system in one homogeneous layer."""
    solution = solve_ivp(
        fun=lambda s, y: jet_equations(s, y, alpha, B_const),
        t_span=(s_start, s_end),
        y0=y0,
        events=events,
        method="RK45",
        rtol=1e-8, # relative tolerance 
        atol=1e-10 # atol (absolute tolerance) 
    )
    return solution


def combine_two_solutions(sol1, sol2):
    """Merge the first and second integration step into one continuous result."""
    combined_s = np.concatenate([sol1.t, sol2.t[1:]])
    combined_y = np.concatenate([sol1.y, sol2.y[:, 1:]], axis=1)

    class CombinedSolution:
        def __init__(self, t, y, status, message):
            self.t = t
            self.y = y
            self.status = status
            self.message = message

    return CombinedSolution(
        t=combined_s,
        y=combined_y,
        status=sol2.status,
        message=sol2.message
    )


def format_result(name, Frd, solution, Q0):
    """Create a dictionary with the final result."""
    return {
        "name": name,
        "Frd": Frd,
        "solution": solution,
        "s": solution.t,
        "Q": solution.y[0],
        "M": solution.y[1],
        "theta_deg": np.rad2deg(solution.y[2]),
        "x": solution.y[3],
        "z": solution.y[4],
        "dilution": solution.y[0] / Q0,
        "status": solution.status,
        "message": solution.message
    }


# =========================================================
# 4. MAIN FUNCTION FOR ONE CASE
# =========================================================
def run_case_two_step(name, alpha, V0, theta0_deg, params, is_stratified=True):
    """
    Solve one case.

    Logic:
    - If the water is stratified:
        Step 1: outlet -> pycnocline
        Step 2: pycnocline -> free surface
    - If the water is mixed:
        Solve directly to the free surface
    """

    # ---- Initial conditions at the source
    Q0, M0, theta0 = calculate_initial_conditions(V0, theta0_deg, params)

    # ---- Reduced gravity in bottom and top layer
    g_prime_bottom = calculate_reduced_gravity(params.rho_bottom, params.rho_0, params.g)
    g_prime_top = calculate_reduced_gravity(params.rho_top, params.rho_0, params.g)

    # ---- Source densimetric Froude number
    Frd = V0 / np.sqrt(abs(g_prime_bottom) * params.D)

    # ---- Event functions
    hit_pycnocline, hit_surface, hit_bottom = create_events(params)

    # =====================================================
    # 1
    # =====================================================
    B_bottom = Q0 * g_prime_bottom
    initial_state = [Q0, M0, theta0, 0.0, params.z_0]

    if is_stratified:
        events_step1 = [hit_pycnocline, hit_bottom]
    else:
        events_step1 = [hit_surface, hit_bottom]

    sol1 = solve_one_layer(
        y0=initial_state,
        alpha=alpha,
        B_const=B_bottom,
        s_start=0.0,
        s_end=500.0,
        events=events_step1
    )

    # If ambient is mixed, or pycnocline is not reached,return only the first solution
    if (not is_stratified) or (sol1.t_events[0].size == 0):
        return format_result(name, Frd, sol1, Q0)

    # =====================================================
    # 2
    # =====================================================
    state_at_pycnocline = sol1.y[:, -1]
    Q_at_pycnocline = state_at_pycnocline[0]

    # Recalculate buoyancy flux in the upper layer
    B_top = Q_at_pycnocline * g_prime_top

    sol2 = solve_one_layer(
        y0=state_at_pycnocline,
        alpha=alpha,
        B_const=B_top,
        s_start=sol1.t[-1],
        s_end=sol1.t[-1] + 500.0,
        events=[hit_surface, hit_bottom]
    )

    # Combine both parts into one continuous result
    combined_solution = combine_two_solutions(sol1, sol2)

    return format_result(name, Frd, combined_solution, Q0)


# =========================================================
# 5. SCENARIOS
# =========================================================
params = OutfallParams()

scenarios = {
    "Base case":     {"alpha": 0.09, "V0": 1.0, "theta0_deg": 0.0,  "strat": True},
    "Low alpha":     {"alpha": 0.07, "V0": 1.0, "theta0_deg": 0.0,  "strat": True},
    "High alpha":    {"alpha": 0.13, "V0": 1.0, "theta0_deg": 0.0,  "strat": True},
    "Angle 45":      {"alpha": 0.09, "V0": 1.0, "theta0_deg": 45.0, "strat": True},
    "Angle 90":      {"alpha": 0.09, "V0": 1.0, "theta0_deg": 90.0, "strat": True},
    "Low velocity":  {"alpha": 0.09, "V0": 0.5, "theta0_deg": 0.0,  "strat": True},
    "High velocity": {"alpha": 0.09, "V0": 2.0, "theta0_deg": 0.0,  "strat": True},
    "Mixed ambient": {"alpha": 0.09, "V0": 1.0, "theta0_deg": 0.0,  "strat": False}
}


# =========================================================
# 6. RUN ALL CASES
# =========================================================
all_results = {}

for case_name, case_params in scenarios.items():
    result = run_case_two_step(
        name=case_name,
        alpha=case_params["alpha"],
        V0=case_params["V0"],
        theta0_deg=case_params["theta0_deg"],
        params=params,
        is_stratified=case_params["strat"]
    )
    all_results[case_name] = result
```
