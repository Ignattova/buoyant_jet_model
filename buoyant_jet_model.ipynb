***Abstract:*** This project investigates the mathematical modeling of key hydrodynamic processes that occur during submarine discharge into coastal waters, where the effluent density differs from that of the ambient water. The primary focus is on the physical processes governing the initial mixing and dilution of the discharged jet, which play a critical role in limiting environmental impacts on marine ecosystems. A one-dimensional (1D) integral plume model is implemented, based on a system of three ordinary differential equations representing the conservation of mass flux (Q), momentum flux (M), and buoyancy flux (B). The model accounts for turbulent entrainment of the ambient water and the effects of vertical density stratification in the seawater, using the Boussinesq approximation. The study provides a theoretical basis for future analysis of the overall mixing processes between wastewater and seawater.

## I. INTRODUCTION

>### I.1 Problem Statement and Motivation
>>Coastal regions hold significant environmental and economic value at both global and national scales. The demand for recreation and tourism has led to a high concentration of population in these areas, resulting in rapid urban expansion and increasing pressure on coastal zones. One of the primary sources of this environmental pressure is the discharge of municipal and industrial wastewater.  
>>*The Bulgarian Black Sea coast* is the main recipient of a significant part of the wastewater in the country - wastewater from urban treatment plants (WWTPs), sewage without complete treatment, as well as significant volumes of urban stormwater runoff. This leads to increased anthropogenic pollution and biogenic nutrient enrichment (eutrophication), which has become a major ecological concern over recent decades. As a result, the entire Bulgarian Black Sea coast has been designated as a ***sensitive area***.


>### I.2 Environmental Quality Standards and Mixing Zones
>>The discharge of wastewater of different composition into water bodies can lead to increased concentrations of chemical substances in the aquatic environment. To mitigate adverse effects on ecosystems and human health, the European Union established the **Water Framework Directive (Directive 2000/60/EC)**, which provides the foundation for integrated water resource management and pollution reduction at the source. Within this regulatory framework, **Environmental Quality Standards (EQS)** were introduced, as defined in **Directives 2008/105/EC and 2013/39/EU**.

>>In the immediate vicinity of a discharge point, concentrations of certain priority substances may temporarily exceed the established standards. The area surrounding the discharge point where such exceedances are permitted is referred to as the **mixing zone**.

>>Delineation of the **mixing zone** is a key element in assessing the environmental impact of wastewater discharges. It allows the spatial variation of pollutant concentrations after wastewater discharge to be analyzed and helps assess the risk of not meeting the environmental objectives established for the receiving water body.

>### I.3.	Mixing Zones in Coastal Waters
>>The mechanisms governing the mixing of wastewater discharges in coastal waters are not uniform throughout the process and vary depending on the distance from the discharge source. Most researchers divide the process into two main regions based on the dominant physical mechanisms: the **near-field** and the **far-field**. Some authors identify a third phase, referred to as a transition region (*Koh & Brooks, 1975; Fischer et al., 1979*), while others consider it to be the final stage of the near-field (*Tsanis et al., 1994*).
>>### I.3.1.	Phase 1: Initial Mixing  
>>This phase begins immediately after the discharge of wastewater into the receiving water body. As the effluent  leaves the diffuser and enters the denser ambient seawater, the jet undergoes **initial dilution**. This stage is characterized by intense turbulent mixing, which leads to rapid dilution of the discharged effluent.The primary driving forces are the initial jet momentum (turbulence generated by the discharge velocity) and the buoyancy-induced flux.  

>>### I.3.2.	Phase 2: Plume Rise and Intermediate Mixing
>>This phase occurs in the *mid-zone* and is characterized by dynamic horizontal dispersion and further vertical dilution. Differences in velocity and density between the effluent and the ambient water induce turbulence as the plume rises. The phase concludes when the plume reaches its terminal rise height. This height may correspond either to the water surface or to a specific depth where the density of the mixture becomes equal to that of the surrounding ambient water. In stratified waters, ambient density stratification reduces vertical mixing, limits the rise height of the jet, and may prevent it from reaching the surface.  
>>The completion of this phase marks the end of the **near-field**, also referred to as the **hydrodynamic mixing zone** or the **initial dilution zone**.

>>### I.3.3.	Phase 3: Far-Field Transport
>>The final phase of wastewater transport is characterized by the spreading of the discharged mixture under the influence of marine currents. The dynamic action of these currents causes dispersion of the mixture, leading to secondary mixing and dilution in the **far-field**. The rate of mixing is primarily controlled by the advection of marine currents and the turbulent characteristics of the ambient water, while the influence of the initial discharge parameters gradually diminishes. In certain cases, sedimentation of suspended particles on the seabed may also occur.


>### I.4.	Scope of the Study
>>The primary focus of this project is placed on the processes occurring within the **near-field region**. The analysis of the near-field is crucial, as it defines the initial conditions for the subsequent transport and dispersion of pollutants in the marine environment. A detailed investigation of these processes provides the basis for the further development of tools that can improve the prediction of potential pollutant concentrations in marine waters, support the optimization of the depth and location of discharge pipelines, and provide a scientifically grounded assessment of whether a given project complies with regulatory water quality standards.
