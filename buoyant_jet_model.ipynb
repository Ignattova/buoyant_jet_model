## Abstract


</div>

This project investigates the mathematical modeling of hydrodynamic processes associated with submarine wastewater discharges in coastal zones. The research focuses on the physical phenomenon of near-field mixing and initial dilution, which is critical for the preservation of marine ecosystems. A one-dimensional (1D) integral model is implemented, based on a system of three ordinary differential equations derived from the conservation of mass (Q), momentum (M), and buoyancy flux (B). The model incorporates the effects of turbulent entrainment and vertical density stratification (thermocline/pycnocline) under the Boussinesq approximation.The governing equations are solved numerically in Python to compute the dilution ratio S(z) and the plume rise trajectory, providing a theoretical foundation for future analysis using machine learning techniques.

## I. Introduction

>### I.1 Problem Statement and Motivation
>>Coastal regions hold significant environmental and economic value at both global and national scales. The demand for recreation and tourism has led to a high concentration of population in these areas, resulting in rapid urban expansion and increasing pressure on the coastline. One of the primary sources of this environmental pressure is the discharge of municipal and industrial wastewater.
 
>> #### The Bulgarian Black Sea Coast

>>The Bulgarian Black Sea coast serves as the main recipient of a large portion of the country's wastewater, including treated effluent from wastewater treatment plants (WWTPs), untreated sewage, and significant volumes of urban surface runoff. This leads to increased anthropogenic pollution and biogenic nutrient enrichment (eutrophication), which has become a major ecological concern over recent decades. As a result, the entire Bulgarian Black Sea coast has been designated as a **sensitive area**.


>### I.2 Water Quality Standards and Mixing Zones

>>The discharge of different types of wastewater into water bodies can lead to increased concentrations of chemical substances in the aquatic environment. To mitigate adverse effects on ecosystems and human health, the European Union established the **Water Framework Directive (Directive 2000/60/EC)**, which provides the foundation for integrated water resource management and pollution reduction at the source. Within this regulatory framework, **Environmental Quality Standards (EQS)** were introduced, as defined in **Directives 2008/105/EC and 2013/39/EU**.

>>In the immediate vicinity of a discharge point, concentrations of certain priority substances may temporarily exceed the established standards. The spatial area surrounding the discharge point where such exceedances are permitted is referred to as the **mixing zone**.

>>Delineation of the **mixing zone** is a key element in assessing the environmental impact of wastewater discharges. It allows the spatial variation of pollutant concentrations to be analyzed after the release of wastewater and helps determine whether there is a risk of failing to meet the environmental objectives established for the receiving water body.
